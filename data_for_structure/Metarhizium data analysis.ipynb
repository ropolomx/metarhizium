## Integration of metadata

### Reading metadata from the Excel file provided by Jenny Gusse

```python
metadata = pd.read_excel('Metadata_Metarhizium_ALL_Sept21_2018.xlsx')
```

```python
metadata.head()
```

```python
metadata.dtypes
```

This is going to be a bit interesting. In the microsatellite data frames from above we have a __SampleID__. In the metadata file provided by Jenny, we have __LRD ID #__ and __Sample ID__, but the __SampleID__ does not seem to match the other __SampleID__ from the microsatellite data. In fact the Sample ID from the metadata is a combination of the __Soil ID__ and the __Sub ID__. That should not be an issue: the first page of the microsatellite file contains both the microsatellite SampleID and the LRC ID. I can merge those, and then merge with the metadata file using the LRC ID.

```python
micro_metadata = pd.read_excel('Microsatellite_data_all_Sept20_2018.xlsx', sheet_name="All isolates")
```

This has been updated to reflect the _M.robertsii_ sample 2702 with LRC ID 293

```python
micro_metadata = micro_metadata.rename(columns={'Sample ID #': 'SampleID'})
```

```python
micro_metadata = micro_metadata.loc[:, lambda df: ['SampleID','LRC ID #']]
```

```python
micro_metadata.head()
```

```python
micro_metadata['SampleID'] = micro_metadata['SampleID'].astype('object')
micro_metadata['LRC ID #'] = micro_metadata['LRC ID #'].astype('object')
```

```python
meta_merge = metadata.merge(micro_metadata, how='inner', on='LRC ID #')
```

```python
#meta_merge['SampleID'] = meta_merge['SampleID'].str.replace('(2828).*$', '\1')
```

```python
meta_merge[meta_merge.SampleID == 2828]
```

```python
meta_merge.shape
```

```python
mrob_meta = microsatellites['Mrob'].merge(meta_merge, how = 'inner', on='SampleID', indicator=True)
```

```python
mrob_meta.shape[0] == microsatellites['Mrob'].shape[0]
```

```python
mrob_meta.head()
```

```python
microsatellites['Mbru'][microsatellites['Mbru'].SampleID == 2828]
```

```python
mbru_meta = microsatellites['Mbru'].merge(meta_merge, how = 'left', on='SampleID', indicator=True)
```

```python
mbru_meta
```

```python
mbru_meta.shape[0] == microsatellites['Mbru'].shape[0]
```

```python
mbru_meta.columns
```

## _M. bru_ with glaciation metadata

```python
mbru_glaciation = mbru_meta.loc[:, lambda df: ['SampleID','Glaciation',
                 'Ma097','Ma325', 'Ma2049', 
                 'Ma375_1', 'Ma375_2', 'Ma375_3',
                 'Ma2060', 'Ma2069', 'Ma307_1', 
                 'Ma2283', 'Ma2296_1', 'Ma2077', 
                 'Ma2063', 'Ma2292', 'Ma2070',
                 'Ma2056','Ma2089','Ma2054']]
```

## Replace "I" with 1 and "O" with 2

```python
mbru_glaciation['Glaciation'] = mbru_glaciation['Glaciation'].str.replace("I","1")
mbru_glaciation['Glaciation'] = mbru_glaciation['Glaciation'].str.replace("O","2")
```
